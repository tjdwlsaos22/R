# Mushroom
- 8124종류의 버섯의 특징과 독의 유무를 정리해놓은 데이터셋
- 버섯 모양의 특징을 기반으로 독의 유무 판단
 
 한 줄(row)이 버섯 한 종류
 1번쨰 컬럼은 독의 유무: 독(p),식용(e)
 2번쨰 컬럼은 버섯의 머리 모양 : b(bell 형태),c(원뿔),x(볼록),f(평평),k(혹),s(오목)
 4번째 컬럼은 버섯의 머리 색: n()

```r
mr <- read.csv("./Data/mushroom.csv",header=F, stringsAsFactors=T)
mr

```

```r
for(i in 2:23){
    mr[,i] = as.numeric(mr[,i])
}
mr
```

```r
library(dplyr)
mr <- arrange(mr,V1)
mr
```

```r
# 정리한 dataset의 e와 p의 갯수확인.
nrow(subset(mr,V1 == "e"))
nrow(subset(mr,V1 == "p"))
nrow(subset(mr,V1 == "p"))/nrow(subset(mr,V1 == "e"))
```

```r
# sampling
library(caret)
set.seed(1234)
samp <- createDataPartition(mr$V1,p=0.7,list=F)
```

```r
data.train <- mr[samp,]
data.test <- mr[-samp,]
```

```r
x <- subset(data.test,select = -V1)
y <- data.test$V1
```

```r
rf <- randomForest(V1~.,data=data.train,ntree = 100)
```

```r
pred <- predict(rf, x)
pred
```

```r
mean(pred==y)
```

```r
table(pred,y)
```

```r
#인공신경망
library(nnet)
set.seed(1234)
samp <- createDataPartition(mr$V1,p=0.7,list=F)
data.tr <- mr[samp,]
data.test <- mr[-samp,]
x <- subset(data.test,select = -V1)
y <- data.test$V1
model.nnet <- nnet(V1~.,data=data.tr,size=3)
pred <-predict(model.nnet,x,type="class") #class는 객관식
head(pred)
mean(pred==y)
table(pred,y)
```

```r
# 로지스틱 회귀분석
set.seed(1234)
samp <- createDataPartition(mr$V1,p=0.7,list=F)

data.tr <- mr[samp,]
data.test <- mr[-samp,]
x <- subset(data.test,select = -V1)
y<- data.test$V1
model <- multinom(V1 ~ .,data=data.tr)
pred <- predict(model,x)
mean(y==pred)
table(pred,y)
```

```r
library(e1071)
samp <-createDataPartition(mr$V1,p=0.7,list=F)

data.tr <- mr[samp,]
data.test <-mr[-samp,]
x <- subset(data.tr,select = -V1)
y <- data.tr$V1
model <- svm(V1~.,data=data.tr)
pred <- predict(model, x)
mean(pred==y)
table(pred,y)
```

# Column이름과 One-Hot Encoding을 한번에 처리하기

```r
mr4 <- data.frame(V1=mr$V1)
```

```r

count = 2 #2번째 컬럼부터 작업 시작후 원본 컬럼번호에 새로운 컬럼이름을  저장하기 위해 사용
for(i in 2:23){
    for(j in sort(unique(mr[,i]))){
        mr4=cbind(mr4, mr[,i]==j)
        colnames(mr4)[count]<-paste("V",i,"_",j,sep="")
        count = count+1
    }
}
head(mr4)
```
