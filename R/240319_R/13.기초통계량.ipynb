### 기초통계량 산출

```r
cafe <-read.csv("./Data/coffee.csv")
head(cafe)
```

```r
#빈도수
table(cafe$Coffee)
```

```r
# 범위 빈도수 (table로 할수없을 정도로 양이 많을때 사용)
coffee_Range <- seq(0,60,10)
table(cut(cafe$Coffee,breaks = coffee_Range))
#0에서 60을 10씩으로 쪼갠것에 대해 빈도수를 측정
#(는 초과  ]는 이하
table(cut(cafe$Coffee,breaks = coffee_Range,right=F))
#(]위치 반대됨.
```

```r
#줄기-잎 그림
stem(cafe$Coffee)
```

```r
#변동계수
# 빈도수, 표준편차
sd(cafe$Coffee)/mean(cafe$Coffee)
sd(cafe$Fruit.Cup)/mean(cafe$Fruit.Cup)

ca.m <- mean(cafe$Coffee)
ca.sd <- sd(cafe$Coffee)

cat("커피 평균 판매량:",round(ca.m,1),"커피 표준편차 :",round(ca.sd,2),"\n")

ju.m <- mean(cafe$Fruit.Cup)
ju.sd <- sd(cafe$Fruit.Cup)

cat("쥬스 평균 판매량:",round(ju.m,1),"쥬스 표준편차 :",round(ju.sd,2))
```

```r
#커피 판매량의 변동계수
cat("커피 변동계수 :",ca.sd/ca.m,"\n")
# 쥬스 판매량의 변동계수
cat("쥬스 변동계수 :",ju.sd/ju.m)
```

```r
sort(cafe$Fruit.Cup)
sort(cafe$Coffee)
```

```r
par(bg='white')
boxplot(cafe$Coffee,cafe$Fruit.Cup)
```
