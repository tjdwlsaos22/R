```r
#모집단에서 필요한 데이터만 추출하기
#iris에서 species가 Setosa인것만 추출해서 iris_Setosa라는 데이터 프레임 만들기

iris_setosa <- subset(iris,Species == 'setosa')
iris_setosa
str(iris_setosa) ## factor가 1개가 아닌 3개가 나오는것이라 바꿔줘야함. setosa만 나와야함. 이거 바형태해도 3개가 나오게 되는 현상나옴

#factor 조정
iris_setosa$Species <- factor(iris_setosa$Species)  # 이렇게 설정해주면 위에 설정한 값만 나오게됨.
str(iris_setosa)
```

```r
#모집단에서 필요한 데이터만 추출하기
#iris에서 species가 versicolor인것만 추출해서 iris_versicolor라는 데이터 프레임 만들기

iris_versicolor <- subset(iris,Species == 'versicolor')
iris_versicolor
str(iris_versicolor) ## factor가 1개가 아닌 3개가 나오는것이라 바꿔줘야함. setosa만 나와야함. 이거 바형태해도 3개가 나오게 되는 현상나옴

#factor 조정
rownames(iris_versicolor) <-c()
head(iris_versicolor)

iris_versicolor$Species <- factor(iris_versicolor$Species)  # 이렇게 설정해주면 위에 설정한 값만 나오게됨.
str(iris_versicolor)
```

```r
#모집단에서 필요한 데이터만 추출하기
#iris에서 species가 setosa이고  Sepal.Length가 5보다 큰 데이터 만들기

iris_setosa1<- subset(iris,Species == 'setosa' & Sepal.Length > 5)
iris_setosa1

str(iris_setosa1) ## factor가 1개가 아닌 3개가 나오는것이라 바꿔줘야함. setosa만 나와야함. 이거 바형태해도 3개가 나오게 되는 현상나옴

rownames(iris_setosa1) <- c()
head(iris_setosa1)

iris_setosa1$Species <- factor(iris_setosa1$Species)  # 이렇게 설정해주면 위에 설정한 값만 나오게됨.

str(iris_setosa1)
head(iris_setosa1)

summary(iris_setosa1)
summary(iris)
```
