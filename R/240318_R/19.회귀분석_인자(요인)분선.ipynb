```r
regression <- read.csv("./Data/regression.csv")
head(regression)

summary(regression)
```

```r
par(bg='white')
hist(regression$Height)
plot(regression$Height,regression$Weight)
```

```r
#상관계수
cor(regression$Height,regression$Weight)
```

```r
#회귀선(중심선,평균선)

par(bg='white')
plot(regression$Height,regression$Weight)

r <- lm(regression$Weight~regression$Height)
abline(r,col='red')
#y =ax+b의 선이 나옴. 위치보고 구할 수 있음.

summary(r)
#residuals = 잔차
#coefficients = 계수 y=ax+b에서 a,b가 계수
```

- p-value: < 2.2e-16 < 0.05  => 분석가능
- Adjusted R-squared:  0.9333 => 예측력
- Residual standard error: 6.355 => 표준오차

```r
#y=ax+b
#몸무게 = 0.6147 *키 + -40.8659 

#키가 170인 사람의 몸무게
0.6147*170+-40.8659

#표준오차를 적용한 키가 170인 사람의 몸무게
cat(0.6147*170+-40.8659-6.355,"~",0.6147*170+-40.8659+6.355)
```

# IQ

```r
iq <- read.csv("./Data/IQ.csv")
```

```r
iq
```

```r
# 성적과 가장 관련이 깊은 항목은?
cor(iq$`성적`,iq$`IQ`)
cor(iq$`성적`,iq$`다니는.학원수`)
cor(iq$`성적`,iq$`게임하는.시간`)
cor(iq$`성적`,iq$`TV시청.시간`)

cor(iq[-1])
```

```r
#iq가 126인 사람의 성적을 예측해라
par(family="AppleGothic",bg='white')
plot(iq$`IQ`,iq$성적)

r <- lm(iq$`성적`~iq$IQ)
abline(r,col='red')

summary(r)


# - p-value: < 0.0002547 < 0.05  => 분석가능
# - Adjusted R-squared:  0.8071  => 예측력
# - Residual standard error: 3.322 => 표준오차
#표준오차를 적용한 iq가 126인 사람의 성적
cat(0.6714 *126 + -5.2918 -3.322,"~",0.6714 *126 + -5.2918  +3.322)
```

```r
# 학원을 몇군데 다녀야 성적이 90점 이상이될까

par(family="AppleGothic",bg='white')
plot(iq$`성적`,iq$`다니는.학원수`)

r <- lm(iq$`다니는.학원수`~iq$`성적`)
abline(r,col='red')

summary(r)


# - p-value: < 0.007264 < 0.05  => 분석가능
# - Adjusted R-squared:  0.5666    => 예측력
# - Residual standard error: 4.98 => 표준오차
#표준오차를 적용한 iq가 126인 사람의 성적
cat(0.12410   *90 + -7.89163 -0.7882,"~",0.12410  *90 + -7.89163  +0.7882)

```

```r


# 성적이 85점이라면 iq는 얼마이상?
par(family="AppleGothic",bg='white')
plot(iq$`성적`,iq$IQ)

r <- lm(iq$IQ~iq$`성적`)
abline(r,col='red')

summary(r)


# - p-value: < 0.0002547 < 0.05  => 분석가능
# - Adjusted R-squared:  0.8071   => 예측력
# - Residual standard error: 4.504 => 표준오차
cat(1.2340   *85 + 28.0353 -4.504,"~",1.2340   *85 + 28.0353  +4.504)
```

# 교통사고 예측하기(회귀식과 RandomForest예측)

```r
seoul <- read.csv("./Data/newSeoul2019.csv")
head(seoul)
```

```r
JongRo <- seoul[seoul$`자치구명` == "종로구", ]
nrow(JongRo)
JongRo
```

```r
tail(JongRo)
rownames(JongRo) <- c(1:108)
JongRo
tail(JongRo)
```

```r
str(JongRo)
```

```r
cor(JongRo$`발생건수`,JongRo$`부상자수`)
```

```r
par(family="AppleGothic",bg='white')
plot(JongRo$`발생건수`,JongRo$`부상자수`,xlab="발생건수",ylab="부상자수",main="종로구의 발생건수와 부상자수의 산포도")

r <- lm(JongRo$`부상자수`~JongRo$`발생건수`)
abline(r,col='red')

summary(r)

```

```r

par(family="AppleGothic",bg='white')
plot(JongRo$`발생건수`,JongRo$`부상자수`)

r <- lm(JongRo$`부상자수`~JongRo$`발생건수`)
abline(r,col='red')

summary(r)


# - p-value: < 2.2e-16 < 0.05  => 분석가능
# - Adjusted R-squared:  0.8466    => 예측력
# - Residual standard error: 9.816 => 표준오차
cat(1.35993*100+2.88612)
```

```r
#랜덤 포레스트 예측 : 발생건수가 100일경우의 부상자수?
library(randomForest)
```

```r
rf <- randomForest(부상자수~발생건수,data=JongRo,ntree = 100)
predict(rf,list(발생건수=100))
```
