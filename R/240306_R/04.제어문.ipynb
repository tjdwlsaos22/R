```r
#excel file에서 q3컬럼의 홀수 합과 짝수 합 구하기
odd=0
even=0
for( i in excel$q3){      
    ifelse(i %% 2 ==0,odd<-odd+i,even<-even+i)
}
cat(odd,"\n")
cat(even)

sum(ifelse(excel$q3 %% 2 ==0,excel$q3,0))
sum(ifelse(excel$q3 %% 2 !=0,excel$q3,0))
#3의 배수도 되고 4의 배수가 되는 것.
summary(excel$q3)
sum(ifelse(excel$q3 %% 3 ==0 && excel$q3 %% 4 ==0 ,excel$q3,0))
```

```r
#구구단 함수 만들기 - 1
gugu <- function(){
    dan <- as.integer(readline("단 입력"))
    for(i in 1:9){
    cat(dan,"X",i,"=",dan*i,"\n")
}
}

gugu()
```

```r

gugu2 <- function(dan){
    for(i in 1:9){
    cat(dan,"X",i,"=",dan*i,"\n")
}
}

gugu2(3)
```

```r
gugu3 <- function(dan,i){
    for(a in 1:i)
    cat(dan,"X",a,"=",dan*a,"\n")
}
gugu3(2,12)
```

```r
# 짝수 단과 짝수 곱해지는 수만 출력
gugu5 <- function() {
  for (dan in seq(2,9,2)) {
    for (i in seq(2,9,2)) {
      cat(dan, "X", i, "=", dan * i, "\n")
    
    }
 cat("--------------","\n") }
}

gugu5()

# 2개의 벡터자료를 함꼐 출력하기
score <- c(85,95,89)
name <- c("유비","관우","장비")

for(i in 1:length(name)){
  cat(name[i],":",score[i],"\n")
}
cat(paste(name,":",score,"\n"),sep="")
cat(name,":",score)

```
