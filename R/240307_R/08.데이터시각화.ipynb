```r
x <- runif(100)
y <- runif(100)

#0.5기준으로 0.5보다 크면 pch=1,아니면 pch=18
#0.5기준으로 0.5보다 크면 red,아니면 blue
par(family = "AppleGothic",bg='lightgray')
#plotting symbol
plot(x,y,
col = ifelse(x >0.5 | y>0.5, 'blue','greenyellow'),
pch=ifelse(x >0.5 | y>0.5, 1,18)
) #계단
```
